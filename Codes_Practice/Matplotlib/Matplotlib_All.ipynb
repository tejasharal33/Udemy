## Matplotlib example Workflow

```python
# 0. import matplotlib and get ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1.prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2.Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # (width * height)

# 3. Plot Data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Two options for subplots
# Subplot 1 
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
#Plot to each different axes
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

### Which one to use pyplot or Matplotlib OO Method

* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method

over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# OO method mix with pyplot
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
#ax.set_xlim([45,100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot data
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

#customize the plot
ax.set(title="Heart Disease and cholestrol levels",
       xlabel="Age",
       ylabel="cholestrol")

#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol,age,thalach

fig, ( ax0, ax1) = plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10,10),
                                sharex=True)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

#Customize ax0
ax0.set(title='heart disease and cholestrol levels',
        ylabel='Cholestrol')

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

#Add a mean line
ax0.axhline(y=over_50["chol"].mean(), linestyle='--')


#Add Data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"]);

#customize ax1
ax1.set(title="Heart Disease and max heart rate", 
        xlabel = "Age",
        ylabel = "Max Heart Rate")

#Add a legend
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a mean line
ax1.axhline(y=over_50["thalach"].mean(), linestyle='--')

#Add a title for figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```

```python
plt.style.use("seaborn-v0_8")

## OO method from scratch 2
fig, ax = plt.subplots(figsize=(10,6))

#plot data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") # changes colour scheme

#customize the plot
ax.set(title="Heart Disease and cholestrol levels",
       xlabel="Age",
       ylabel="cholestrol")

#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
## Customizing the y and x axes limitation

# Subplot of chol,age,thalach

fig, ( ax0, ax1) = plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10,10),
                                sharex=True)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

#Customize ax0
ax0.set(title='heart disease and cholestrol levels',
        ylabel='Cholestrol')

# Change x axis limits
ax0.set_xlim([50,80])

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

#Add a mean line
ax0.axhline(y=over_50["chol"].mean(), linestyle='--')


#Add Data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter");

#customize ax1
ax1.set(title="Heart Disease and max heart rate", 
        xlabel = "Age",
        ylabel = "Max Heart Rate")

# Change x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#Add a legend
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a mean line
ax1.axhline(y=over_50["thalach"].mean(), linestyle='--')

#Add a title for figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```
