## Matplotlib example Workflow

```python
# 0. import matplotlib and get ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1.prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2.Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # (width * height)

# 3. Plot Data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

#### there can be multiple axes on a figure and can share single x or y axis some times

## Making fig with NumPy arrays

Type :
* Line
* Scatter

```python
import numpy as np
```

```python
#create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
x
```

```python
# plot the data and create line plot (default)
fig, ax = plt.subplots()
ax.plot(x,x**2)
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Another scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# make plot from dictionary
nut_butter_prices = {"Almond butter":10, 
                     "Peanut Butter":8,
                     "Cashew Butter":12}

fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values()) #x,y
ax.set(title = "Tejas's Nut butter store",
       ylabel="Prices ($)",); # use ; at last line of ax or plot
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),nut_butter_prices.values())
```

```python
#data

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Two options for subplots
# Subplot 1 
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
#Plot to each different axes
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

#Plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```
